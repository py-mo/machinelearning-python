```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Import data
data = pd.read_csv("../preprocessing/data.csv")
data.head()
```

```python
wo_data = pd.read_csv("../preprocessing/wo_data.csv")
wo_data.head()
```

```python
# Create K-means model
for n in range(3, 11):
    print (f"------------------{n}------------------")
    km_model = KMeans(n_clusters = n, n_init = 20)
    km_model.fit(data.values)
    print('Silhoutte score of our model is ' + str(silhouette_score(data, km_model.labels_)))
```

```python
# Create K-means model
for n in range(3, 11):
    print (f"------------------{n}------------------")
    wo_km_model = KMeans(n_clusters = n, n_init = 20)
    wo_km_model.fit(wo_data.values)
    print('Silhoutte score of our model is ' + str(silhouette_score(wo_data, wo_km_model.labels_)))
```

We should divide customers to some groups
If we divide them in three groups(good, normal, bad) we have better accuracy and also it can be useful

```python
km_model = KMeans(n_clusters = 3, n_init = 20)
km_model.fit(data.values)
print('Silhoutte score of our model is ' + str(silhouette_score(data, km_model.labels_)))
```
